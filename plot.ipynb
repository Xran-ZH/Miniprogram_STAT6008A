```python
import matplotlib.pyplot as plt
import numpy as np

def a_0(x):
    if x <= 0:
        exit(1)
    elif 0<x<=1:
        return 0.9*(1-np.exp(-x))
    elif 1<x<=2:
        return 0.9*(np.exp(1-x)-np.exp(-x)+0.5*np.exp(1-2*x)-0.5/np.e)
    else:
        return 0.9*(np.exp(1-x)-np.exp(-x)+0.5*np.exp(1-2*x)-0.5*np.exp(3-2*x))
    
def a_1(x):
    if x <= 0:
        exit(1)
    elif 0<x<=1:
        return 0
    elif 1<x<=2:
        return 1/np.e-np.exp(-x)
    else:
        return np.exp(1-x)-np.exp(-x)
plt.rc('text', usetex=True)
plt.rc('font', family='sans-serif')
plt.rcParams["font.family"] = ["Times New Roman"]
ls=26
fs=30
lw=3
colors = ['#E4A031', '#D68438', '#C76B60', '#B55384', '#7C4D77', '#474769', 
          '#B2B6C1', '#73A5A2', '#4C7780', '#F6E2C1', '#F3DBC1', '#D6E2E2', 
          '#F0EFED', '#E3BBED', '#26445E']
x = np.linspace(0, 6, 10000)
y0 = [a_0(xi) for xi in x]
y1 = [a_1(xi) for xi in x]
plt.figure(figsize=(10,5))
plt.plot(x, y0, label=r'$\psi(x,a_0)$', color=colors[0],linewidth=lw)
plt.plot(x, y1, label=r'$\psi(x,a_1)$', color=colors[8],linewidth=lw)
plt.scatter([1.65127404], [a_0(1.651274036)], color=colors[-1],s=70, zorder=10,label=r'$x_0$')
plt.xlabel(r'$x$',fontsize=fs)
plt.ylabel('Function values',fontsize=fs)
#plt.title(r'Plots of $\psi(x,a_0)$ and $\psi(x,a_1)$')
plt.tick_params(direction='in', top=True, right=True,labelsize=ls)
plt.legend(loc='best',fontsize=ls)
plt.grid()
plt.savefig('psi_functions.pdf', bbox_inches='tight', pad_inches=0,dpi=200,transparent=True)
```
